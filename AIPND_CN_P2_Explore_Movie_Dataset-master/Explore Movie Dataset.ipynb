**任务2.3: **分组读取

1. 对 `release_year` 进行分组，使用 [`.agg`](http://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.core.groupby.DataFrameGroupBy.agg.html) 获得 `revenue` 的均值。
2. 对 `director` 进行分组，使用 [`.agg`](http://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.core.groupby.DataFrameGroupBy.agg.html) 获得 `popularity` 的均值，从高到低排列。

要求：使用 `Groupby` 命令实现。

```python
release_year_gb = movie_data.groupby("release_year")
print(release_year_gb.agg({"revenue":"mean"}))
```

```python
director_gb = movie_data.groupby("director")

print(director_gb.agg({"popularity":"mean"}).sort_values(by = "popularity", ascending = False))
```

**任务3.1：**对 `popularity` 最高的20名电影绘制其 `popularity` 值。

```python
movie_data_popularity_20 = movie_data.sort_values(by = "popularity", ascending = False)[0:20]

base_color = sb.color_palette()[0]

sb.barplot(data = movie_data_popularity_20, x = "original_title",y = "popularity", color = base_color);
plt.xticks(rotation = 90)
```

---
**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
movie_data["net_margin"] = movie_data["revenue"] - movie_data["budget"]

ax = sb.lineplot(data = movie_data, x = "release_year", y = "net_margin")
ax.set_xlabel("release_year")
ax.set_ylabel("net_margin")
```
